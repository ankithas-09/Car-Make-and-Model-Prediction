```python
def load_image(img_path, input_shape, resize=False):
    if resize:
        img, _ = resize_black(input_shape[1], img_path, print_oldsize=False)
    else:
        img = image.load_img(img_path, target_size=input_shape)
    img_tensor = image.img_to_array(img)
    img_tensor = np.expand_dims(img_tensor, axis=0)
    img_tensor /= 255.
    return img_tensor

def generate_data_generators(input_shape):
    return create_data_generators(input_shape, BATCHSIZE, TRAIN_DIR, TEST_DIR, save_augmented=None, plot_imgs=False)

def load_model(results_folder):
    with open(os.path.join(results_folder, 'model.json'), 'r') as json_file:
        loaded_model_json = json_file.read()
    loaded_model = model_from_json(loaded_model_json)
    loaded_model.load_weights(os.path.join(results_folder, "weights.best.hdf5"))
    input_shape = loaded_model.layers[0].output_shape[1:3]
    print("Loaded model from disk")
    return loaded_model, input_shape

def print_confusion_matrix(cls_pred, cls_test, class_names, cmap=plt.cm.Blues):
    cm = confusion_matrix(y_true=cls_test, y_pred=cls_pred)
    print("Confusion matrix:")
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.colorbar()
    plt.xlabel('Predicted label')
    plt.ylabel('True label')
    plt.xticks(np.arange(len(class_names)), class_names, rotation=45)
    plt.yticks(np.arange(len(class_names)), class_names)
    plt.show()

def decode_predictions(preds, class_names, top=5):
    results = []
    for pred in preds:
        top_indices = pred.argsort()[-top:][::-1]
        result = [(class_names[i], pred[i]) for i in top_indices]
        result.sort(key=lambda x: x[1], reverse=True)
        results.append(result)
    return results

def predict(img_path, model, input_shape, class_names, correct_class, resize=False, show_imgs=True, top=5):
    img_array = load_image(img_path, input_shape, resize)
    preds = model.predict(img_array)
    predictions = decode_predictions(preds, class_names, top=top)
    top1_pred = predictions[0][0]

    if show_imgs:
        img_org = image.load_img(img_path)
        fig, axs = plt.subplots(1, 2)
        axs[0].imshow(img_org)
        axs[0].set_title(f"Actual: {correct_class}")
        axs[1].imshow(img_array[0])
        axs[1].set_title(f"Predicted: {top1_pred[0]}")
        plt.show()

    return predictions

def show_model_performance(loaded_model, generator_train, generator_test):
    steps_test = generator_test.n / BATCHSIZE
    cls_train = generator_train.classes
    cls_test = generator_test.classes
    class_names = list(generator_train.class_indices.keys())
    y_pred = loaded_model.predict(generator_test, steps=steps_test)
    cls_pred = np.argmax(y_pred, axis=1)
    print_confusion_matrix(cls_pred, cls_test, class_names)

def return_class_names_list(results_folder):
    with open(os.path.join(results_folder, 'class_names.txt'), 'r') as file:
        return [line.strip() for line in file]

def perform_pred(loaded_model, resize=False, car_class=None, img_pth=None, show_imgs=True):
    test_dir = TEST_DIR_TST
    results_folder = RESULTS_FOLDER

    if os.path.exists(os.path.join(results_folder, 'class_names.txt')):
        class_names = return_class_names_list(results_folder)
    else:
        generator_train, _ = generate_data_generators(input_shape)
        class_names = list(generator_train.class_indices.keys())
        with open(os.path.join(results_folder, 'class_names.txt'), 'w') as file:
            for name in class_names:
                file.write(f'{name}\n')

    if car_class is None:
        car_class = random.choice(class_names)

    if img_pth is None:
        test_img = os.path.join(test_dir, car_class, random.choice(os.listdir(os.path.join(test_dir, car_class))))
    else:
        test_img = img_pth

    input_shape = loaded_model.layers[0].output_shape[1:3]
    predictions = predict(test_img, loaded_model, input_shape, class_names, car_class, resize=resize, show_imgs=show_imgs)
    print(predictions)
```

```python
def extract_make(class_name):
    return class_name.split(' ', 1)[0]

def extract_cartype(class_name):
    elements = class_name.split()
    return elements[-3] if class_name == 'Infiniti G Coupe IPL 2012' else elements[-2]

def get_most_common_brands(class_names, top_n=30):
    makes = [extract_make(name) for name in class_names]
    most_common_makes = Counter(makes).most_common(top_n)
    return [make for make, _ in most_common_makes]

def get_most_common_car_types(class_names, top_n=9):
    car_types = [extract_cartype(name) for name in class_names]
    most_common_car_types = Counter(car_types).most_common(top_n)
    return [car_type for car_type, _ in most_common_car_types]

def is_popular_brand(class_name, popular_brands):
    return extract_make(class_name) in popular_brands

def is_popular_type(class_name, popular_types):
    return extract_cartype(class_name) in popular_types
```

```python
def configure_ax(ax):
    """Configure plot axes for a cleaner look."""
    for spine in ax.spines.values():
        spine.set_visible(False)
    ax.set_ylabel('')
    ax.set_xlabel('')
    ax.yaxis.set_label_fontsize(12)
    ax.xaxis.set_label_fontsize(12)
    plt.title(ax.get_title(), fontsize=14, weight='bold')

def save_and_show_plot(title, savepath):
    """Save plot to file and show it."""
    if savepath:
        fig_path = f"{savepath}/{title}.png"
        plt.savefig(fig_path)
    plt.show()

def plot_horizontal_bar(df, title='', fontsize=12, savepath='', many_data=True):
    figsize = (10, 25) if many_data else (10, 7)
    ax = df.plot.barh(color='#86bf91', zorder=2, width=0.85, figsize=figsize)
    configure_ax(ax)
    ax.set_ylabel('Odpowiedź sieci', fontsize=fontsize, weight='bold')
    ax.set_xlabel('Wystąpienia', fontsize=fontsize, weight='bold')
    plt.title(title, fontsize=fontsize + 2, weight='bold')
    save_and_show_plot(title, savepath)

def plot_accuracy_for_all_classes(class_and_acc, title='', fontsize=12, savepath='', many_data=False):
    df = pd.DataFrame(list(class_and_acc.items()), columns=['Class', 'Accuracy'])
    df.sort_values('Accuracy', ascending=True, inplace=True)
    df.set_index('Class', inplace=True)
    
    figsize = (10, 25) if many_data else (10, 7)
    ax = df.plot.barh(color='#86bf91', zorder=2, width=0.85, figsize=figsize)
    configure_ax(ax)
    ax.set_ylabel('Klasy', fontsize=fontsize, weight='bold')
    ax.set_xlabel('Dokładność', fontsize=fontsize, weight='bold')
    plt.title(title, fontsize=fontsize + 2, weight='bold')
    save_and_show_plot(title, savepath)

def mostcommon_predictions(list_of_preds, title='', fontsize=12, savepath='', many_data=True):
    cnt = Counter(list_of_preds)
    df = pd.DataFrame(cnt.items(), columns=['Prediction', 'Count'])
    df.sort_values('Count', ascending=True, inplace=True)
    df.set_index('Prediction', inplace=True)
    
    figsize = (10, 25) if many_data else (10, 7)
    ax = df.plot.barh(color='#86bf91', zorder=2, width=0.85, figsize=figsize, fontsize=fontsize)
    configure_ax(ax)
    ax.set_ylabel('Odpowiedź sieci', fontsize=fontsize, weight='bold')
    ax.set_xlabel('Wystąpienia', fontsize=fontsize, weight='bold')
    plt.title(title, fontsize=fontsize + 2, weight='bold')
    save_and_show_plot(title, savepath)
    return cnt

def plot_prediction_analysis_all(metrics):
    labels = ['Niepoprawne wskazania', 'Poprawne wskazania', 'Niepoprawne wskazania - ale zgadza się marka lub typ']
    sizes = metrics[3], metrics[0], metrics[2]
    colors = ['yellowgreen', 'lightcoral', 'lightskyblue']
    explode = (0, 0.1, 0)
    
    patches, texts, autotexts = plt.pie(sizes, explode=explode, colors=colors,
                                        autopct='%1.1f%%', shadow=True, startangle=160)
    plt.legend(patches, labels, loc="best")
    plt.tight_layout()
    plt.axis('equal')
    plt.show()

def plot_prediction_analysis_brand(metrics):
    correct_brand = metrics[0] + metrics[1]
    incorrect_brand = (sum(metrics) - metrics[1]) - correct_brand
    
    labels = ['Prawidłowa marka', 'Nieprawidłowa marka']
    sizes = [correct_brand, incorrect_brand]
    colors = ['gold', 'lightskyblue']
    explode = (0.1, 0)
    
    patches, texts, autotexts = plt.pie(sizes, explode=explode, colors=colors,
                                        autopct='%1.1f%%', shadow=True, startangle=160)
    plt.legend(patches, labels, loc="best")
    plt.tight_layout()
    plt.axis('equal')
    plt.show()
```

```python
# Helper functions
def configure_ax(ax):
    for spine in ax.spines.values():
        spine.set_visible(False)
    ax.set_ylabel('')
    ax.set_xlabel('')
    plt.title(ax.get_title(), fontsize=14, weight='bold')

def save_and_show_plot(title, savepath):
    if savepath:
        fig_path = f"{savepath}/{title}.png"
        plt.savefig(fig_path)
    plt.show()

def plot_horizontal_bar(df, title='', fontsize=12, savepath='', many_data=True):
    figsize = (10, 25) if many_data else (10, 7)
    ax = df.plot.barh(color='#86bf91', zorder=2, width=0.85, figsize=figsize)
    configure_ax(ax)
    ax.set_ylabel('Odpowiedź sieci', fontsize=fontsize, weight='bold')
    ax.set_xlabel('Wystąpienia', fontsize=fontsize, weight='bold')
    plt.title(title, fontsize=fontsize + 2, weight='bold')
    save_and_show_plot(title, savepath)

def plot_accuracy_for_all_classes(class_and_acc, title='', fontsize=12, savepath='', many_data=False):
    df = pd.DataFrame(list(class_and_acc.items()), columns=['Class', 'Accuracy'])
    df.sort_values('Accuracy', ascending=True, inplace=True)
    df.set_index('Class', inplace=True)
    
    figsize = (10, 25) if many_data else (10, 7)
    ax = df.plot.barh(color='#86bf91', zorder=2, width=0.85, figsize=figsize)
    configure_ax(ax)
    ax.set_ylabel('Klasy', fontsize=fontsize, weight='bold')
    ax.set_xlabel('Dokładność', fontsize=fontsize, weight='bold')
    plt.title(title, fontsize=fontsize + 2, weight='bold')
    save_and_show_plot(title, savepath)

def mostcommon_predictions(list_of_preds, title='', fontsize=12, savepath='', many_data=True):
    cnt = Counter(list_of_preds)
    df = pd.DataFrame(cnt.items(), columns=['Prediction', 'Count'])
    df.sort_values('Count', ascending=True, inplace=True)
    df.set_index('Prediction', inplace=True)
    
    figsize = (10, 25) if many_data else (10, 7)
    ax = df.plot.barh(color='#86bf91', zorder=2, width=0.85, figsize=figsize, fontsize=fontsize)
    configure_ax(ax)
    ax.set_ylabel('Odpowiedź sieci', fontsize=fontsize, weight='bold')
    ax.set_xlabel('Wystąpienia', fontsize=fontsize, weight='bold')
    plt.title(title, fontsize=fontsize + 2, weight='bold')
    save_and_show_plot(title, savepath)
    return cnt

def plot_prediction_analysis_all(metrics):
    labels = ['Niepoprawne wskazania', 'Poprawne wskazania', 'Niepoprawne wskazania - ale zgadza się marka lub typ']
    sizes = [metrics[3], metrics[0], metrics[2]]
    colors = ['yellowgreen', 'lightcoral', 'lightskyblue']
    explode = (0, 0.1, 0)
    
    patches, texts, autotexts = plt.pie(sizes, explode=explode, colors=colors,
                                        autopct='%1.1f%%', shadow=True, startangle=160)
    plt.legend(patches, labels, loc="best")
    plt.tight_layout()
    plt.axis('equal')
    plt.show()

def plot_prediction_analysis_brand(metrics):
    correct_brand = metrics[0] + metrics[1]
    incorrect_brand = (sum(metrics) - metrics[1]) - correct_brand
    
    labels = ['Prawidłowa marka', 'Nieprawidłowa marka']
    sizes = [correct_brand, incorrect_brand]
    colors = ['gold', 'lightskyblue']
    explode = (0.1, 0)
    
    patches, texts, autotexts = plt.pie(sizes, explode=explode, colors=colors,
                                        autopct='%1.1f%%', shadow=True, startangle=160)
    plt.legend(patches, labels, loc="best")
    plt.tight_layout()
    plt.axis('equal')
    plt.show()

def predict_on_class(model, input_shape, class_names, common_brands, common_types, folder_path, resize=False, threshold=0.9999, show_plots=True, print_analysis=True, savepath=''):
    class_name = os.path.basename(folder_path)
    class_name_brand = take_make_from_classname(class_name)
    class_name_cartype = take_cartype_from_classname(class_name)
    
    n_correct_brand = 0 if is_popular_brand(class_name, common_brands) else None
    n_correct_brand_or_type = 0 if is_popular_type(class_name, common_types) else None
    n_totally_incorrect = 0 if is_popular_brand(class_name, common_brands) else None
    
    if n_correct_brand is None:
        print("The brand of car is not a popular brand. Statistics concerning brand won't be provided.")
    if n_correct_brand_or_type is None:
        print("The type of car is not a popular type. Statistics concerning type won't be provided.")

    main_return_dict = {}
    n_correct_make_and_model = 0
    n_corrects_high_confidence = 0
    n_incorrect_high_confidence = 0
    n_all_samples = 0
    incorrect_predictions_high_confidence = []
    incorrect_predictions = []
    all_top1_predictions = []
    top5accuracy = 0
    top5_predictions = []
    top5_predictions_high = []

    for img in os.scandir(folder_path):
        n_all_samples += 1
        predictions = predict(os.path.abspath(img), model, input_shape, class_names, class_name, resize=resize, top=1, show_imgs=False)
        top_prediction = predictions[0][0]
        all_top1_predictions.append(top_prediction[0])
        
        if top_prediction[0] == class_name:
            n_correct_make_and_model += 1
            if float(top_prediction[1]) >= threshold:
                n_corrects_high_confidence += 1
        else:
            incorrect_predictions.append(top_prediction[0])
            if float(top_prediction[1]) >= threshold:
                n_incorrect_high_confidence += 1
            if n_correct_brand is not None:
                if take_make_from_classname(top_prediction[0]) == class_name_brand:
                    n_correct_brand += 1
                    n_correct_brand_or_type += 1
                elif n_correct_brand_or_type is not None and take_cartype_from_classname(top_prediction[0]) == class_name_cartype:
                    n_correct_brand_or_type += 1
            elif n_correct_brand_or_type is not None and take_cartype_from_classname(top_prediction[0]) == class_name_cartype:
                n_correct_brand_or_type += 1

        if n_correct_brand_or_type is not None:
            n_totally_incorrect = len(incorrect_predictions) - n_correct_brand_or_type
        else:
            n_correct_brand_or_type = len(incorrect_predictions)
    
    accuracy = n_correct_make_and_model / n_all_samples * 100
    correctly_pred_high_conf_perc = n_corrects_high_confidence / n_all_samples * 100
    incorrectly_pred_high_conf_perc = n_incorrect_high_confidence / n_all_samples * 100
    brand_type_analysis = [n_correct_make_and_model, n_correct_brand, n_correct_brand_or_type, n_totally_incorrect]
    
    if print_analysis:
        print(f"All samples: {n_all_samples}. Correctly predicted: {n_correct_make_and_model}. Accuracy: {accuracy:.2f}%.")
        print(f"Correctly predicted with high confidence: {n_corrects_high_confidence} => {correctly_pred_high_conf_perc:.2f}% of all.")
        print(f"Incorrectly predicted with high confidence: {n_incorrect_high_confidence} => {incorrectly_pred_hih_conf_perc:.2f}% of all.")
    
    if show_plots:
        if n_correct_brand is not None:
            plot_prediction_analysis_all(brand_type_analysis)
            plot_prediction_analysis_brand(brand_type_analysis)
        elif n_correct_brand_or_type is not None:
            plot_prediction_analysis_all(brand_type_analysis)
        
        mostcommon_predictions(all_top1_predictions, class_name, many_data=False, savepath=savepath)
    
    return accuracy, correctly_pred_high_conf_perc, incorrectly_pred_hih_conf_perc, brand_type_analysis
```
